```python
def find_length_of_LCIS(nums):
    if not nums:
        return 0

    max_length = 1
    current_length = 1

    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 1

    return max_length

# Example usage:
nums = [1, 3, 5, 4, 7]
print("Output:", find_length_of_LCIS(nums))  # Output: 3
```

```python

from functools import cmp_to_key

def largest_number(nums):
    # Custom comparator function to compare two numbers
    def compare(x, y):
        return int(y + x) - int(x + y)

    # Convert numbers to strings for comparison
    nums_str = [str(num) for num in nums]

    # Sort the numbers using the custom comparator
    nums_str.sort(key=cmp_to_key(compare))

    # Handle the case when the largest number is 0
    if nums_str[0] == '0':
        return '0'

    # Concatenate the sorted numbers
    return ''.join(nums_str)

# Example 1
nums1 = [10, 2]
print("Example 1 Output:", largest_number(nums1))

# Example 2
nums2 = [3, 30, 34, 5, 9]
print("Example 2 Output:", largest_number(nums2))
```
